```python
### Extra : None Homework

## Get Data -> Filter Variable -> Grouping -> Reset DataFrame -> Model -> Analysis

import os
import numpy as np
import pandas as pd

# 設定 data_path, 並讀取 app_train
dir_data = 'D:/GitHub/Data/Data1'
f_app_train = os.path.join(dir_data, 'AI_data_2.csv')
app_train = pd.read_csv(f_app_train)

app_train.head()


## Load Data
sub_train = pd.DataFrame(app_train ,columns=['SHIP_DATE','CUST_COUNTRY','Qty'] )
print(sub_train.shape)
sub_train


## Grouping Data(eliminate none use vb )
sub_train_groupby = sub_train.groupby(['SHIP_DATE','CUST_COUNTRY'])
dictData = {}
Month = []
Country = []
Qty = []
for i in sub_train_groupby:
    Month.append(i[0][0])
    Country.append(i[0][1])
    Qty.append(i[1]["Qty"].values[0])
    print(i[0][0], i[0][1], i[1]["Qty"].values[0] )

    
## Reset Colimns
dictData["Month"] = Month
dictData["Country"] = Country
dictData["Qty"] = Qty

new_train = pd.DataFrame(dictData )
new_train     



## numerical vb
sub_train_OneHowEncoding = pd.get_dummies(new_train ,columns=['Country']) 

sub_train_OneHowEncoding
```
