#### 請寫一個函式用來計算 Mean Square Error
$ MSE = \frac{1}{n}\sum_{i=1}^{n}{(Y_i - \hat{Y}_i)^2} $

### Hint: [如何取平方](https://googoodesign.gitbooks.io/-ezpython/unit-1.html)

```python
import numpy as np
import matplotlib.pyplot as plt
import math


w = 3
b = 0.5

x_lin = np.linspace(0, 100, 101)
x_lin = np.array(x_lin,  dtype=np.int32 )

y = ((x_lin + (np.random.randn(101) * 5)) * w) + b
y = np.array(y,  dtype=np.int32 )
print(y)
print(y**2)
 
plt.plot(x_lin, y, 'b.', label = 'data points')
plt.title("Assume we have data points")
plt.legend(loc = 2)
plt.show()

y_hat = x_lin * w + b
y_hat = np.array(y_hat,dtype=np.int32 )
print(y_hat)
print(y_hat**2)

plt.plot(x_lin, y, 'b.', label = 'data')
plt.plot(x_lin, y_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()

def mean_squared_error(ActualArrany, PredictArray):  
    MainArray =(ActualArrany - PredictArray)
    MainArray = np.array(MainArray,  dtype=np.int32 )
    #print(ActualArrany)
    #print(PredictArray)
    #print(MainArray)
    
    MainArray = MainArray**2
    #print(MainArray)
    
    #print(MainArray.sum() )
    ArrayCounter = np.size(MainArray)
    #print(ArrayCounter)
    MSE = MainArray.sum() / ArrayCounter

    return MSE

def mean_absolute_error(ActualArrany, PredictArray):  
    MainArray = abs(ActualArrany - PredictArray)
    MainArray = np.array(MainArray,  dtype=np.int32 )
    ArrayCounter = np.size(MainArray)
    MAE = MainArray.sum() / ArrayCounter
  
    return MAE

# 執行 Function, 確認有沒有正常執行 
MSE = mean_squared_error(y, y_hat)
MAE = mean_absolute_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))
print("The Mean absolute error is %.3f" % (MAE))
```
